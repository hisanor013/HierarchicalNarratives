# Gexf Func

```python
import collections

def WriteGexf(nodes,attrS,attrST,attrD,attrDT,edges,hsize,ofile):
    import collections
    # Convert attribute
    attr2id = {}
    cnt = 0
    if(len(attrS)>0):
        for i in range(len(attrS)):
            attr2id.update({attrS[i] : cnt})
            cnt += 1
    if(len(attrD)>0):
        for i in range(len(attrD)):
            attr2id.update({attrD[i] : cnt})
            cnt += 1  
    
    gexf0 = """<?xml version="1.0" encoding="UTF-8"?>
    <gexf xmlns="http://www.gexf.net/1.2draft" 
    xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" xsi:schemaLocation="http://www.gexf.net/1.2draft http://www.gexf.net/1.2draft/gexf.xsd" xmlns:viz="http://www.gexf.net/1.2draft/viz" version="1.2">
    <meta lastmodifieddate="2009-03-20">
    <creator>{}</creator>
    <description>{}</description>
    </meta>
    <graph mode="dynamic" defaultedgetype="directed" timeformat="date">
    """.format("RH","Dynamic Network")

    gexf1 = """<attributes class="node" mode="static">"""
    for i in range(len(attrS)):
        gexf1 = gexf1 + "\n" + "<attribute id=\"" + str(attr2id[attrS[i]]) + \
        "\" title=\"" + attrS[i] + "\" type=\"" + attrST[i] + "\"/>"
    gexf1 = gexf1 + "\n" + "</attributes>\n"

    gexf2 = """<attributes class="node" mode="dynamic">"""
    for i in range(len(attrD)):
        gexf2 = gexf2 + "\n" + "<attribute id=\"" + str(attr2id[attrD[i]]) + \
        "\" title=\"" + attrD[i] + "\" type=\"" + attrDT[i] + "\"/>\n"
    gexf2 = gexf2 + "</attributes>\n"
    
    gexf3 = """<nodes>"""
    for i in range(len(nodes)):
        
        #print(nodes[i])
        gexf3 = gexf3 + "\n" + "<node id=\"" + str(i) + \
            "\" label=\"" + nodes[i]["name"] + \
            "\" start=\"" + nodes[i]["start"] + \
            "\" end=\"" + nodes[i]["end"] + "\">"
        gexf3 = gexf3 + "\n<attvalues>\n"
        for j in range(len(attrS)):
            gexf3 = gexf3 +  "<attvalue for=\"" + str(attr2id[attrS[j]]) + \
                "\" value=\"" + \
                nodes[i][attrS[j]] + "\"/>\n"

        for j in range(len(attrD)):
            if attrD[j] in nodes[i]:
                gexf3 = gexf3 +  "<attvalue for=\"" + str(attr2id[attrD[j]]) + \
                    "\" value=\"" + \
                    str(hsize) + "\" start=\"" + nodes[i][attrD[j]]["start"] + "\" end=\"" + \
                    nodes[i][attrD[j]]["end"] + "\"/>\n"
            else:
                gexf3 = gexf3 +  "<attvalue for=\"" + str(attr2id[attrD[j]]) + \
                    "\" value=\"" + \
                    "0\" start=\"" + "2000-01-01" + "\" end=\"" + \
                    "2001-01-01" + "\"/>\n"
                    
        gexf3 = gexf3 + "</attvalues>\n"
                    
        # if we want to alter shape
        if "shape" in nodes[i]:
            gexf3 = gexf3 + "<viz:shape value=\"" + nodes[i]["shape"] + \
                "\" />\n"
            
        # if there is a spell
        if "spellSt" in nodes[i]:
            NumSpell = len(nodes[i]["spellSt"])
            gexf3 = gexf3 + "<spells>\n"
            for k in range(NumSpell):
                gexf3 = gexf3 + "<spell start=\"" + nodes[i]["spellSt"][k] + \
                    "\" end=\"" + nodes[i]["spellEn"][k] + "\" />\n"
            gexf3 = gexf3 + "</spells>\n"

        gexf3 = gexf3 + "</node>\n"
    gexf3 = gexf3 + "</nodes>\n"
    
    gexf4 = "<edges>\n"
    for i in range(len(edges)):
        gexf4 = gexf4 + "<edge id=\"" + str(i) + \
        "\" source=\"" + str(edges[i]["source"]) + \
        "\" target=\"" + str(edges[i]["target"]) + \
        "\" weight=\"" + str(edges[i]["weight"]) + \
        "\" start=\"" + str(edges[i]["start"]) + \
        "\" end=\"" + str(edges[i]["end"]) + "\"/>\n"
    gexf4 = gexf4 + "</edges>\n"
    gexf4 = gexf4 + "</graph>\n</gexf>\n"
    gexf = gexf0 + gexf1 + gexf2 + gexf3 + gexf4

    file = open(ofile,"w") 
    file.write(gexf) 
    file.close()
```

# Create Node and Edges

```python
import pandas as pd
net_id = 15
dfE = pd.read_csv("log_public/df_edge.csv",dtype=str)
dfN = pd.read_csv("log_public/df_node.csv",dtype=str)
df_sent = pd.read_csv("log_public/sentiment_gpt.csv",dtype=str)
print(len(dfE))
dfE = dfE.dropna()
print(len(dfE))

#dfE["weight"] = 1.0
node2label = dict()
for i in range(len(dfN)):
    node2label.update({dfN["Id"].iloc[i]:dfN["label"].iloc[i]})
    
# Convert the 'date' column to datetime format
dfE['start'] = pd.to_datetime(dfE['date'])

# Add one year to the 'date' column
dfE['end'] = dfE['start'] + pd.DateOffset(years=1)
dfE['start'] = dfE['start'].dt.strftime('%Y-%m-%d')
dfE['end'] = dfE['end'].dt.strftime('%Y-%m-%d')
```

```python
ofile = "log_public/public_opinion" + ".gexf"
hsize = 12
attrS = ["random"]
attrST = ["string"]
attrD  = ["on"]
attrDT = ["float"]

## Nodes
UniNodes = list(dfE["source"])
UniNodes.extend(list(dfE["target"]))
UniNodes = list(set(UniNodes))
UniNodes.sort()

Date = list(dfE["start"])
Date.extend(list(dfE["end"]))
Date = list(set(Date))
Date.sort()
Start = Date[0]
End = Date[-1]
```

```python
# Nodes
nodes2id = {}
nodes = collections.defaultdict(dict)
for i in range(len(UniNodes)):
    nodes2id.update({UniNodes[i] : i})
    condA = dfE["source"] == UniNodes[i]
    dfA = dfE.loc[condA]
    condB = dfE["target"] == UniNodes[i]
    dfB = dfE.loc[condB]
    tDate = list(set(dfA["start"]))
    tDate.extend(list(set(dfA["end"])))
    tDate.extend(list(set(dfB["start"])))
    tDate.extend(list(set(dfB["end"])))
    tDate = list(set(tDate))
    tDate.sort()
    
    
    if tDate[0] != "2000-01-01":
        print(tDate[0])
        print(UniNodes[i])
    
    # show only when have interactions
    nodes[i]["start"] = tDate[0]
    nodes[i]["end"] = tDate[-1]
    # show along
    #nodes[i]["start"] = Date[0]
    #nodes[i]["end"] = Date[-1]    
    #nodes[i]["name"] = UniNodes[i]
    nodes[i]["name"] = node2label[UniNodes[i]]
    nodes[i]["random"] = "1"

    condC = dfN["Id"] == UniNodes[i]
    if sum(condC) > 0:
        if False:
            dftemp = dfN.loc[condC]
            on = {"start" : dftemp["start"].iloc[0], "end" : dftemp["end"].iloc[0]}
            nodes[i]["on"] = on

    # shape either disc,square,triangle,diamond
    #nodes[0]["shape"] = "triangle"
    #spellSt = {0 : "2000-01-01", 1: "2000-03-01"}
    #spellEn = {0 : "2000-02-01", 1: "2001-01-01"}
    #nodes[0]["spellSt"] = spellSt
    #nodes[0]["spellEn"] = spellEn

    
# Edges
edges = collections.defaultdict(dict)
for i in range(len(dfE)):
    edges[i]["source"] = nodes2id[dfE["source"].iloc[i]]
    edges[i]["target"] = nodes2id[dfE["target"].iloc[i]]
    edges[i]["start"] = dfE["start"].iloc[i]
    edges[i]["end"] = dfE["end"].iloc[i]
    edges[i]["weight"] = 1000.0*float(dfE["weight"].iloc[0])

```

# Write to files

```python
%%time
WriteGexf(nodes,attrS,attrST,attrD,attrDT,edges,hsize,ofile)
```
