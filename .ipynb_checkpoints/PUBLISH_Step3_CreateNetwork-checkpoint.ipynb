# Read lines

````python
#def read_log(path):
index2hantei = dict()
    
with open(path, "r") as f:
    lines = f.readlines()

out_info,out_edges = [],[]
num_read = 0
error_cnt = 0
for kkk in range(len(lines)):
    data = json.loads(lines[kkk])
    index = data["index"]
    if index not in index2hantei:
        input_text = data["input"]
        answer = data["answer"]
          
        try:
            if answer.startswith('```json'):
                # Strip the markdown code block indicators
                answer = answer.strip('```json\\n')
                # Load the inner JSON
                answer_dict = json.loads(answer)

            elif "\n}\n}" in answer:
                answer = answer.replace("\n}\n}","]\n}")
                answer_dict = json.loads(answer)
            else:
                answer_dict = json.loads(answer)
            num_read += 1
            answer_type = answer_dict['種類']
            A = answer_dict["A"]
            B = answer_dict["B"]
            C = answer_dict["C"]
            D = answer_dict["D"]
            E = answer_dict["E"]
            F = answer_dict["F"]
            summary = answer_dict['要約文']
            out_edges.append([index,input_text,answer_type,A,B,C,D,E,F,summary])
            index2hantei.update({index:1})
        except:
            error_cnt += 1
                
print("Number of error: " + str(error_cnt))
````

````python
df_out = pd.DataFrame(out_edges)
df_out.columns = ["index","input","type","A","B","C","D","E","F","summary"]
print("Number of error: " + str(error_cnt))

nodes = list(set(df_out["A"].dropna()))
nodes.extend(list(set(df_out["A"].dropna())))
nodes.extend(list(set(df_out["B"].dropna())))
nodes.extend(list(set(df_out["C"].dropna())))
nodes.extend(list(set(df_out["D"].dropna())))
nodes.extend(list(set(df_out["E"].dropna())))
nodes.extend(list(set(df_out["F"].dropna())))
nodes = list(set(nodes))
````

# Embedding

````python
model_embed = "use"
````

````python
# THIS MODEL MIGHT NOT BE SUITABLE FOR PYTHON 3.10 and LATER
if model_embed == "use":
    
    import tensorflow
    import tensorflow_hub as hub
    import tensorflow_text
    embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder-multilingual-large/3")
    
else:
    
    from transformers import BertModel, BertTokenizer
    import torch

    def load_model_and_tokenizer():
        # Load tokenizer and model from Hugging Face Hub
        tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
        model = BertModel.from_pretrained('bert-base-multilingual-cased')
        return tokenizer, model

    def get_embeddings(texts, tokenizer, model):
        # Tokenize input texts and prepare input tensors
        encoded_input = tokenizer(texts, return_tensors='pt', padding=True, truncation=True, max_length=128)
        # Move tensors to the same device as the model
        input_ids = encoded_input['input_ids']

        with torch.no_grad():  # Do not compute gradients
            # Forward pass, get hidden states
            output = model(input_ids)
            # We take the embeddings from the last hidden state
            embeddings = output.last_hidden_state[:, 0, :]
        return embeddings
    tokenizer, model = load_model_and_tokenizer()
    # Get embeddings
````

# Restrict to largest connected component

````python
import pandas as pd
import networkx as nx

def largest_connected_component(edgelist):
    """
    Finds the largest connected component from an edge list DataFrame.

    Parameters:
    edgelist (pd.DataFrame): DataFrame containing the edges of the graph with columns 'source', 'target', and 'weight'.

    Returns:
    pd.DataFrame: DataFrame containing the edges of the largest connected component.
    """
    # Create a graph from the DataFrame
    G = nx.from_pandas_edgelist(edgelist, 'source', 'target', edge_attr='weight')
    
    # Find all connected components, sorted by size
    largest_component = max(nx.connected_components(G), key=len)
    
    # Create a subgraph of the largest component
    subgraph = G.subgraph(largest_component).copy()
    
    # Convert the subgraph back to a DataFrame
    largest_component_df = nx.to_pandas_edgelist(subgraph)
    
    return largest_component_df
````
